```python
import pandas as pd
import numpy as np
from numpy import percentile
import altair as alt
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep=';')
```

```python
print('Quality Mean: %.3f' % df['quality'].mean())
```

```python
quartiles = percentile(df['quality'], [25, 50, 75])
data_min, data_max = df['quality'].min(), df['quality'].max()
print('Quality Min: %.3f' % data_min)
print('Quality Q1: %.3f' % quartiles[0])
print('Quality Median: %.3f' % quartiles[1])
print('Quality Q3: %.3f' % quartiles[2])
print('Quality Max: %.3f' % data_max)
```

```python
(alt.Chart(df).mark_bar().encode(
    alt.X('quality', bin=alt.Bin(maxbins=10), scale=alt.Scale(domain=(1, 10))),
    y='count()'))
```

```python
rows = list(df.columns)

print(rows)
```

```python
corr_df = df.select_dtypes('number').corr('spearman').stack().reset_index(name='corr')
corr_df.loc[corr_df['corr'] == 1, 'corr'] = 0
corr_df['abs'] = corr_df['corr'].abs()
alt.Chart(corr_df).mark_circle().encode(
    x='level_0',
    y='level_1',
    size='abs',
    color=alt.Color('corr', scale=alt.Scale(scheme='blueorange')))
```

```python
(alt.Chart(df)
 .mark_point(size=2, opacity=0.2).encode(
     alt.X(alt.repeat('row'), type='quantitative'),
     alt.Y('quality', type='quantitative', scale=alt.Scale(domain=(1, 10))))
 .properties(width=200, height=200)
 .repeat(row=rows))
```
